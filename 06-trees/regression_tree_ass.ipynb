```python
df = pd.read_csv('car_fuel_efficiency.csv')
df.head()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
column_list = list(df.columns)
column_list
```

```python
for c in column_list:
    if df[c].isna().sum() > 0 and df[c].dtypes == 'object':
        df[c] = df[c].fillna('NA')
    elif df[c].isna().sum() > 0 and df[c].dtypes != 'object':
        df[c] = df[c].fillna(0)
    else:
        pass

df.isna().sum()
```

```python
from sklearn.model_selection import train_test_split

df_full_train, df_test = train_test_split(df, test_size=0.2, random_state=1)
df_train, df_val = train_test_split(df_full_train, test_size=0.25, random_state=1)
```

```python
df_train.head(3)
```

```python
df_train = df_train.reset_index(drop=True)
df_val = df_val.reset_index(drop=True)
df_test = df_test.reset_index(drop=True)
```

```python
y_train = df_train.fuel_efficiency_mpg.values
y_val = df_val.fuel_efficiency_mpg.values
y_test = df_test.fuel_efficiency_mpg.values
```

```python
df_train.head(2)
```

Delete the target variable from the train/val/test dataframe.

```python
del df_train['fuel_efficiency_mpg']
del df_val['fuel_efficiency_mpg']
del df_test['fuel_efficiency_mpg']
```

```python
df_train.head(2)
```

```python
df_val.isnull().sum()
```

```python
from sklearn.feature_extraction import DictVectorizer
```

```python
dv = DictVectorizer(sparse=False)
```

```python
train_dicts = df_train.to_dict(orient='records')
X_train = dv.fit_transform(train_dicts)
```

```python
dv.get_feature_names_out()
```

```python
val_dicts = df_val.to_dict(orient='records')
X_val = dv.transform(val_dicts)
```

Let's train a decision tree regressor to predict the fuel_efficiency_mpg variable.

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dtr = DecisionTreeRegressor(max_depth=1)
dtr.fit(X_train, y_train)
```

```python
list(dv.get_feature_names_out())
```

```python
list(dtr.feature_importances_)
```

Train a random forest regressor with these parameters:

n_estimators=10
random_state=1
n_jobs=-1 (optional - to make training faster)

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
```

```python
rfr = RandomForestRegressor(n_estimators=10, random_state=1, n_jobs=-1)
rfr.fit(X_train, y_train)
```

```python
y_pred = rfr.predict(X_val)
mse = mean_squared_error(y_val, y_pred)
rmse = np.sqrt(mse)
```

```python
rmse
print("Root Mean Squared Error: %s" % (rmse))
```

```python
for n in range(10, 201, 10):
    rfr = RandomForestRegressor(n_estimators=n, random_state=1)
    rfr.fit(X_train, y_train)

    y_pred = rfr.predict(X_val)
    mse = mean_squared_error(y_val, y_pred)
    rmse = np.sqrt(mse)

    print("Root Mean Squared Error: %.3f for %a" % (rmse, n))
```

```python
for depth in [10, 15, 20, 25]:
    for n in range(10, 201, 10):
        rfr = RandomForestRegressor(n_estimators=n, random_state=1, max_depth=depth)
        rfr.fit(X_train, y_train)

        y_pred = rfr.predict(X_val)
        mse = mean_squared_error(y_val, y_pred)
        rmse = np.sqrt(mse)

        print("Depth = %d and Root Mean Squared Error: %.3f for %a" % (depth ,rmse, n))
```

```python

rfr = RandomForestRegressor(n_estimators=10, random_state=1, max_depth=20, n_jobs=-1)
rfr.fit(X_train, y_train)

y_pred = rfr.predict(X_val)
mse = mean_squared_error(y_val, y_pred)
rmse = np.sqrt(mse)

print("Root Mean Squared Error: %.3f" % (rmse))
```

```python
import xgboost as xgb
```

```python
features = list(dv.get_feature_names_out())
dtrain = xgb.DMatrix(X_train, label=y_train, feature_names=features)
dval = xgb.DMatrix(X_val, label=y_val, feature_names=features)
```

```python
watchlist = [(dtrain, 'train'), (dval, 'val')]
```

```python
xgb_params = {
    'eta': 0.1, 
    'max_depth': 6,
    'min_child_weight': 1,
    
    'objective': 'reg:squarederror',
    'nthread': 8,
    
    'seed': 1,
    'verbosity': 1,
}
```

```python
model = xgb.train(xgb_params, dtrain, num_boost_round=100, evals=watchlist)
```

```python
y_pred = model.predict(dval)
mse = mean_squared_error(y_val, y_pred)
rmse = np.sqrt(mse)
print("Root Mean Squared Error: %.3f for %a" % (rmse, xgb_params['eta']))
```

```python
y_pred = model.predict(dval)
mse = mean_squared_error(y_val, y_pred)
rmse = np.sqrt(mse)
print("Root Mean Squared Error: %.3f for %a" % (rmse, xgb_params['eta']))
```
